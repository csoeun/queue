Подключение модели

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')
```

Обучение модели на датасете

```python
results = model.train(
   data='/content/data.yaml',
   imgsz=(1048, 586),
   epochs=15,
   batch=8,
   name='yolov8n_v8_50e'
)
```
